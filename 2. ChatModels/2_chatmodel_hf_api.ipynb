```python
from dotenv import load_dotenv
import os

load_dotenv()
print("Token visible:", os.getenv("HUGGINGFACEHUB_API_TOKEN") is not None)
```

```python
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from dotenv import load_dotenv

load_dotenv()

# switch to a reliable, supported model on the free tier
llm = HuggingFaceEndpoint(
    repo_id="HuggingFaceH4/zephyr-7b-beta", # Much smarter and reliable than TinyLlama
    task="text-generation",
    temperature=0.5,
    max_new_tokens=512
)

model = ChatHuggingFace(llm=llm)

result = model.invoke("What is the capital of United States?")

print(result.content)
```

```python
from langchain_huggingface import ChatHuggingFace, HuggingFacePipeline

# Change 1: Set device to -1 (CPU) to prevent the crash
llm = HuggingFacePipeline.from_model_id(
    model_id="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    task="text-generation",
    device=-1,  # <--- FORCE CPU HERE
    pipeline_kwargs={"temperature":0.7, "max_new_tokens":256}
)
```

```python
from langchain_huggingface import ChatHuggingFace, HuggingFacePipeline

llm = HuggingFacePipeline.from_model_id(
    model_id="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    task="text-generation",
    device=-1,  # <--- FORCE CPU HERE
    pipeline_kwargs={"temperature":0.7, "max_new_tokens":256}
)

model = ChatHuggingFace(llm=llm)

result = model.invoke("What is the capital of United States?")

print(result.content)
```
